```python
spb_df = pd.read_csv('spb_df.csv', index_col=0)
spb_df = spb_df[spb_df.price>10000]
X, y = spb_df.drop(columns=['price', 'address', 'material', 'lat', 'lng', 'data_new', 'id']), spb_df['price']
```

```python
encoder = ce.CatBoostEncoder()
X_cb = encoder.fit_transform(X, y)
```

```python
X = X_cb
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
modellgbm = LGBMRegressor(**lgbm_param).fit(X_train, y_train)
modelxg = XGBRegressor(**xg_param).fit(X_train, y_train)
modelcb = CatBoostRegressor(**cb_param).fit(X_train, y_train)
```

```python
mean_absolute_percentage_error (y, (modellgbm.predict(X)+modelxg.predict(X)+modelcb.predict(X))/3)
```
